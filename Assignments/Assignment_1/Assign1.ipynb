```python
def buildTree(S, vol, T, N):
    dt = T/N

    matrix = np.zeros((N+1, N+1))

    u = np.exp(vol*np.sqrt(dt))
    d = np.exp(-vol*np.sqrt(dt))
    
    # Set stock price to be the first element
    matrix[0,0] = S

    # Iterate over the lower triangle, at this point first element still get erased, need to fix
    for i in np.arange(N + 1): # iterate over rows
        for j in np.arange(i + 1): # iterate over the columns


                # Hint: express each cell as a combination of up and down moves
                matrix[i,j] = S * u**j * d**(i-j)

    return matrix
```

```python
def valueOption(stock_price, strike_price, option_type, position):
    """Values an option given a stock price, strike price, type of option either 'call' or 'put' and a position: 'long' or 'short'"""
    if option_type == 'call':
        # If call option, payoff equals
        pay_off = np.maximum(0, stock_price - strike_price)
    elif option_type == 'put':
        # If put option, payoff equals
        pay_off = np.maximum(0, strike_price - stock_price)
    # If the position is short, return minus pay off
    if position == 'short':
        return -1*pay_off
    # If the position is long, return pay off
    else:
        return pay_off
```

```python
def Calculate_delta(u, d, S0, K, option_type='call', position = 'long'):
    """"Returns a delta calculated based on the option value in up state minus down state devided by the stock value in up state minus down state. 
    Takes parameters u, d, stock price S0, Strike price K, optio_type (set to call by default) and positon (set to long by default)
    Returns delta"""
    f_u = valueOption(stock_price= S0*u, strike_price= K, option_type=option_type, position=position)
    f_d = valueOption(stock_price= S0*d, strike_price= K, option_type=option_type, position=position)
    delta = (f_u - f_d)/(S0*u - S0*d)
    return delta
```

```python
def BlackScholes(S0, St, K, r, T, t, vol, mode = 'call'):
    """"Returns option valuation according to the black and scholes method"""
    d1 = (np.log(St/K) + (r + vol**2/2)*(T-t))/vol*np.sqrt(T-t)
    d2 = d1 - vol*np.sqrt(T-t)
    if mode == 'call':
        option_price = S0*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    elif mode == 'put':
        option_price = K*np.exp(-r*T)*norm.cdf(-d2) - S0*norm.cdf(-d1)
    return option_price, d1, d2
```

```python
def valueOptionMatrix(N, tree, T, r, K, vol, option_type= 'call', position = 'long'):
    dt = T / N
    u = np.exp(vol*np.sqrt(dt))
    d = np.exp(-vol*np.sqrt(dt))
    p = (np.exp(r*dt) - d)/(u - d)


    columns = tree.shape[1]
    rows = tree.shape[0]

    # Walk backward, we start in last row of the matrix

    if option_type == 'call':

    # Add the payoff function in the last row
        for c in np.arange(columns):
            S = tree[rows - 1, c] # value in the matrix
            tree[rows - 1, c] = valueOption(stock_price= S, strike_price=K, option_type=option_type, position = 'long')

        # For all other rows, we need to combine from previous rows
        # We walk backwards, from the last row to the first row
        for i in np.arange(rows - 1)[::-1]:
            for j in np.arange(i + 1):
                down = tree[i + 1, j]
                up = tree[i + 1, j + 1]
                tree[i, j] = np.exp(-r*dt) * (p*up + (1-p)*down)
    return tree

    
```

```python
import numpy as np

def american_option_binomial_tree(S, K, r, T, sigma, option_type='call', steps=1000):
    dt = T / steps
    u = np.exp(sigma * np.sqrt(dt))
    d = 1 / u
    p = (np.exp(r * dt) - d) / (u - d)

    # Initialize stock price tree
    stock_price_tree = np.zeros((steps + 1, steps + 1))
    for i in range(steps + 1):
        for j in range(i + 1):
            stock_price_tree[j, i] = S * (u ** (i - j)) * (d ** j)

    # Initialize option value tree
    option_value_tree = np.zeros((steps + 1, steps + 1))
    if option_type == 'call':
        option_value_tree[:, steps] = np.maximum(0, stock_price_tree[:, steps] - K)
    elif option_type == 'put':
        option_value_tree[:, steps] = np.maximum(0, K - stock_price_tree[:, steps])

    # Calculate option prices at each node
    for i in range(steps - 1, -1, -1):
        for j in range(i + 1):
            if option_type == 'call':
                option_value_tree[j, i] = np.maximum(stock_price_tree[j, i] - K,
                                                      np.exp(-r * dt) * (p * option_value_tree[j, i + 1] +
                                                                         (1 - p) * option_value_tree[j + 1, i + 1]))
            elif option_type == 'put':
                option_value_tree[j, i] = np.maximum(K - stock_price_tree[j, i],
                                                      np.exp(-r * dt) * (p * option_value_tree[j, i + 1] +
                                                                         (1 - p) * option_value_tree[j + 1, i + 1]))

    return option_value_tree[0, 0]

# Example usage:
S = 100  # Current stock price
K = 100  # Strike price
r = 0.05  # Risk-free interest rate
T = 1.0  # Time to expiration (in years)
sigma = 0.2  # Volatility
option_type = 'call'  # 'call' or 'put'
steps = 1000  # Number of steps in the binomial tree

american_option_price = american_option_binomial_tree(S, K, r, T, sigma, option_type, steps)
print("American option price:", american_option_price)
```

```python
def blackScholes(current_stock_price, strike_price, current_time, maturity_time, interest_rate, volatality):
    
    tao = maturity_time - current_time
    d1 = (np.log(current_stock_price/strike_price) + (interest_rate + volatality**2/2) * tao)/(volatality*np.sqrt(tao))
    d2 = d1 - volatality * np.sqrt(tao)
    option_price = current_stock_price * norm.cdf(d1) - np.exp(-interest_rate * tao) * strike_price * norm.cdf(d2)
    
    return option_price, norm.cdf(d1)
```

```python
def exact_Method(S0, interest_rate, delta_t, M, volatality):
    final_stock = []
    for _ in range(100):
        prev_S = S0
        stock_prices = []
        for m in range(1, M+1):
            Zm = np.random.normal()
            next_S = prev_S * np.exp((interest_rate - (volatality)**2 / 2)* delta_t + volatality * np.sqrt(delta_t) * Zm)
            prev_S = next_S
            stock_prices.append(next_S)
        final_stock.append(np.array(stock_prices))
    return (np.mean(final_stock, axis=0))
```

```python
def euler_Method(S0, interest_rate, delta_t, M, volatality):
    final_stock = []
    for _ in range(1):
        prev_S = S0
        stock_prices = [S0]
        for m in range(1, M):
            Zm = norm.rvs()
            next_S = prev_S * ( 1 + interest_rate * delta_t + volatality * np.sqrt(delta_t) * Zm )
            prev_S = next_S
            stock_prices.append(next_S)
        final_stock.append(np.array(stock_prices))
    return (np.mean(final_stock, axis=0))
```

```python
def calculate_Delta_Option(E_stock_prices, strike_price, T , interest_rate, volatality):
    deltas = []
    option_values = []
    time_values = np.array(list(range(1, M+1)))/M
    for i in range(len(time_values) - 1):
        option_value, delta = blackScholes(E_stock_prices[i], strike_price, time_values[i], T, interest_rate, volatality)
        option_values.append(option_value)
        deltas.append(delta)
    return option_values, deltas
```
