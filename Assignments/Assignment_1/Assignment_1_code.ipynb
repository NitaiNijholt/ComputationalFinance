# Plotting Pay-off functions of options

```python
# Params
strike_price = 50
premium_call = 0
premium_put = 0
stock_prices = np.linspace(30,70,100)

# Calculate payoffs

# Note when we go long we pay the premium
payoff_long_call = np.maximum(0, stock_prices - strike_price) - premium_call
payoff_long_put = np.maximum(0, strike_price - stock_prices) - premium_put
# When we go short we receive the premium
payoff_short_call = -np.maximum(0,stock_prices - strike_price) + premium_call
payoff_short_put = -np.maximum(0, strike_price - stock_prices) + premium_put
bond_payoff_value = strike_price

# Plotting
plt.figure(figsize=(14,7))

# Payoff diagrams:

# Long Call
plt.subplot(3, 2, 1)
plt.plot(stock_prices, payoff_long_call, color='blue', label='Long Call Payoff')
plt.axhline(0, color='black', lw=1)
plt.axvline(strike_price, color='red', linestyle='--', lw=1, label='Strike Price')
plt.title('Long Call Option Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)

# Long Put
plt.subplot(3, 2, 2)
plt.plot(stock_prices, payoff_long_put, color='blue', label='Long Put Payoff')
plt.axhline(0, color='black', lw=1)
plt.axvline(strike_price, color='red', linestyle='--', lw=1, label='Strike Price')
plt.title('Long Put Option Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)

# Short Call
plt.subplot(3, 2, 3)
plt.plot(stock_prices, payoff_short_call, color='blue', label='Short Call Payoff')
plt.axhline(0, color='black', lw=1)
plt.axvline(strike_price, color='red', linestyle='--', lw=1, label='Strike Price')
plt.title('Short Call Option Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)

# Short Put 
plt.subplot(3, 2, 4)
plt.plot(stock_prices, payoff_short_put, color='blue', label='Short Put Payoff')
plt.axhline(0, color='black', lw=1)
plt.axvline(strike_price, color='red', linestyle='--', lw=1, label='Strike Price')
plt.title('Short Put Option Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)

# Payoff stock
plt.subplot(3, 2, 5)
plt.plot(stock_prices, stock_prices - strike_price, color='blue', label='Stock Payoff')
plt.axhline(0, color='black', lw=1)
plt.axvline(strike_price, color='red', linestyle='--', lw=1, label='Strike Price')
plt.title('Stock Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)

# Payoff bonds
plt.subplot(3, 2, 6)
plt.axhline(bond_payoff_value, color='blue', linestyle='-', lw=1, label='Bond Payoff')
plt.axvline(strike_price, color='red', linestyle='--', lw=1, label='Strike Price')
plt.title('Bond Payoff')
plt.xlabel('Stock Price at Expiration')
plt.ylabel('Payoff')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
def valueOption(stock_price, strike_price, option_type, position):
    """Values an option given a stock price, strike price, type of option either 'call' or 'put' and a position: 'long' or 'short'"""
    if option_type == 'call':
        # If call option, payoff equals
        pay_off = np.maximum(0, stock_price - strike_price)
    elif option_type == 'put':
        # If put option, payoff equals
        pay_off = np.maximum(0, strike_price - stock_price)
    # If the position is short, return minus pay off
    if position == 'short':
        return -1*pay_off
    # If the position is long, return pay off
    else:
        return pay_off
```

```python
def Calculate_delta(u, d, S0, K, option_type='call', position = 'long'):
    """"Returns a delta calculated based on the option value in up state minus down state devided by the stock value in up state minus down state. 
    Takes parameters u, d, stock price S0, Strike price K, optio_type (set to call by default) and positon (set to long by default)
    Returns delta"""
    f_u = valueOption(stock_price= S0*u, strike_price= K, option_type=option_type, position=position)
    f_d = valueOption(stock_price= S0*d, strike_price= K, option_type=option_type, position=position)
    delta = (f_u - f_d)/(S0*u - S0*d)
    return delta
```

```python
def BlackScholes(S0, St, K, r, T, t, vol, mode = 'call'):
    """"Returns option valuation according to the black and scholes method"""
    d1 = (np.log(St/K) + (r + vol**2/2)*(T-t))/vol*np.sqrt(T-t)
    d2 = d1 - vol*np.sqrt(T-t)
    if mode == 'call':
        option_price = S0*norm.cdf(d1) - K*np.exp(-r*T)*norm.cdf(d2)
    elif mode == 'put':
        option_price = K*np.exp(-r*T)*norm.cdf(-d2) - S0*norm.cdf(-d1)
    return option_price
```

```python
def valueOptionMatrix(N, tree, T, r, K, vol, option_type= 'call', position = 'long'):
    dt = T / N
    u = np.exp(vol*np.sqrt(dt))
    d = np.exp(-vol*np.sqrt(dt))
    p = (np.exp(r*dt) - d)/(u - d)


    columns = tree.shape[1]
    rows = tree.shape[0]

    # Walk backward, we start in last row of the matrix

    if option_type == 'call':

    # Add the payoff function in the last row
        for c in np.arange(columns):
            S = tree[rows - 1, c] # value in the matrix
            tree[rows - 1, c] = valueOption(stock_price= S, strike_price=K, option_type=option_type, position = 'long')

        # For all other rows, we need to combine from previous rows
        # We walk backwards, from the last row to the first row
        for i in np.arange(rows - 1)[::-1]:
            for j in np.arange(i + 1):
                down = tree[i + 1, j]
                up = tree[i + 1, j + 1]
                tree[i, j] = np.exp(-r*dt) * (p*up + (1-p)*down)
    return tree

    
```
