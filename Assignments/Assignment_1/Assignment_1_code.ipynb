```python
def buildTree(S, vol, T, N):
    dt = T/N

    matrix = np.zeros((N+1, N+1))

    u = np.exp(vol*np.sqrt(dt))
    d = np.exp(-vol*np.sqrt(dt))
    
    # Set stock price to be the first element
    matrix[0,0] = S

    # Iterate over the lower triangle, at this point first element still get erased, need to fix
    for i in np.arange(N + 1): # iterate over rows
        for j in np.arange(i + 1): # iterate over the columns


                # Hint: express each cell as a combination of up and down moves
                matrix[i,j] = S * u**j * d**(i-j)

    return matrix
```

```python
def valueOption(stock_price, strike_price, option_type, position):
    """Values an option given a stock price, strike price, type of option either 'call' or 'put' and a position: 'long' or 'short'"""
    if option_type == 'call':
        # If call option, payoff equals
        pay_off = np.maximum(0, stock_price - strike_price)
    elif option_type == 'put':
        # If put option, payoff equals
        pay_off = np.maximum(0, strike_price - stock_price)
    # If the position is short, return minus pay off
    if position == 'short':
        return -1*pay_off
    # If the position is long, return pay off
    else:
        return pay_off
```

```python
def Calculate_delta(u, d, S0, K, option_type='call', position = 'long'):
    """"Returns a delta calculated based on the option value in up state minus down state devided by the stock value in up state minus down state. 
    Takes parameters u, d, stock price S0, Strike price K, optio_type (set to call by default) and positon (set to long by default)
    Returns delta"""
    f_u = valueOption(stock_price= S0*u, strike_price= K, option_type=option_type, position=position)
    f_d = valueOption(stock_price= S0*d, strike_price= K, option_type=option_type, position=position)
    delta = (f_u - f_d)/(S0*u - S0*d)
    return delta
```

```python
def BlackScholes(St, K, r, T, t, vol, mode='call'):
    """Returns option valuation according to the Black-Scholes method."""
    d1 = (np.log(St / K) + (r + 0.5 * vol ** 2) * (T - t)) / (vol * np.sqrt(T - t))
    d2 = d1 - vol * np.sqrt(T - t)
    
    if mode == 'call':
        option_price = St * norm.cdf(d1) - K * np.exp(-r * (T - t)) * norm.cdf(d2)
    elif mode == 'put':
        option_price = K * np.exp(-r * (T - t)) * norm.cdf(-d2) - St * norm.cdf(-d1)
    
    return option_price, norm.cdf(d1), norm.cdf(d2)
```

```python
def valueOptionMatrix(N, tree, T, r, K, vol, option_type= 'call', position = 'long', excise_type= 'european'):
    dt = T / N
    u = np.exp(vol*np.sqrt(dt))
    d = np.exp(-vol*np.sqrt(dt))
    p = (np.exp(r*dt) - d)/(u - d)


    columns = tree.shape[1]
    rows = tree.shape[0]

    # Walk backward, we start in last row of the matrix


# Add the payoff function in the last row
    for c in np.arange(columns):
        S = tree[rows - 1, c] # value in the matrix
        tree[rows - 1, c] = valueOption(stock_price= S, strike_price=K, option_type=option_type, position = 'long')


    # For all other rows, we need to combine from previous rows
    # We walk backwards, from the last row to the first row
    for i in np.arange(rows - 1)[::-1]:
        for j in np.arange(i + 1):
            down = tree[i + 1, j]
            up = tree[i + 1, j + 1]
            continuation_value = np.exp(-r*dt) * (p*up + (1-p)*down)
            if excise_type == 'american':
                intrinsic_value = valueOption(stock_price=tree[i, j], strike_price=K, option_type=option_type, position=position)
                tree[i, j] = max(intrinsic_value, continuation_value)
            else:
                tree[i, j] = continuation_value
    return tree

    
```

```python
# Check if put call parity holds
def put_call_parity(C, P, K, S0, T, r, t=None):
    if t is not None:
        delta_t = T - t
    else:
        delta_t = T
    # Check if put-call parity holds
    if np.isclose(C + K * np.exp(-r * delta_t), P + S0):
        print(f'Given Call price {C} and Put price {P}, the put-call parity holds.')
    else:
        print('Put-call parity does not hold.')
```

```python
K = 99
r = 0.06
St = 100
T = 1
N_range = np.arange(1, 75)  # Range of N values

# Define multiple volatility levels to include in the plot
volatility_levels = [0.1, 0.2, 0.4]  # Example volatility levels

# Prepare plot for Call Option Pricing
plt.figure(figsize=(8, 4))
for vol in volatility_levels:
    # Calculate the analytical option price
    optionPriceAnalytical = BlackScholes(St=St, K=K, r=r, T=T, t=0, vol=vol)[0]

    approximatedPrices = []  # List to store approximated prices for different N
    # Calculate option price for each N using the Binomial Tree model
    for n in N_range:
        treeN = buildTree(S=St, vol=vol, T=T, N=n)
        optionMatrix = valueOptionMatrix(tree=treeN, T=T, r=r, K=K, vol=vol, option_type='call', position='long', N=n)
        priceApproximately = optionMatrix[0, 0]
        approximatedPrices.append(priceApproximately)
    
    # Plot the analytical price as a dashed line and the approximated prices
    plt.plot(N_range, [optionPriceAnalytical]*len(N_range), label=f'BS Price, σ = {vol}', linestyle='--')
    plt.plot(N_range, approximatedPrices, label=f'Binomial Price, σ = {vol}')

plt.xlabel('Number of Steps in Binomial Tree (N)')
plt.ylabel('Call Option Price')
plt.title('Call Option Pricing: Black-Scholes vs. Binomial Tree Model')
plt.legend()
plt.show()

# Prepare plot for Absolute Error in Call Option Pricing
plt.figure(figsize=(8, 4))
for vol in volatility_levels:
    # Re-calculate the analytical option price
    optionPriceAnalytical = BlackScholes(St=St, K=K, r=r, T=T, t=0, vol=vol)[0]

    errors = []  # List to store errors for different N
    # Calculate the error for each N
    for n in N_range:
        treeN = buildTree(S=St, vol=vol, T=T, N=n)
        optionMatrix = valueOptionMatrix(tree=treeN, T=T, r=r, K=K, vol=vol, option_type='call', position='long', N=n)
        priceApproximately = optionMatrix[0, 0]
        error = abs(priceApproximately - optionPriceAnalytical)
        errors.append(error)

    # Plot the errors
    plt.plot(N_range, errors, label=f'Error, σ = {vol}')

plt.xlabel('Number of Steps in Binomial Tree (N)')
plt.ylabel('Absolute Error')
plt.title('Absolute Error in Call Option Pricing as N increases')
plt.legend()
plt.show()
```

The deviation from the analytical value is higher for higher volatility values. Also convergence takes longer for higher volatility values

```python
def exact_GBM(S0, M, T, sigma, r):
    """Returns an analyitical aproximation of the geometric brownian timeseries with volatility sigma of M + 1 steps over a period of T given a starting stock price S0, according to thr analaytical aproximation given in assignment 1 appendix C."""
    delta_t = T/M
    S = np.zeros(M+1)
    S[0] = S0
    np.random.seed(0)
    for t in range(1, M+1):
        Z_m = norm.rvs(0,1)
        S[t] = S[t-1]*np.exp((r - 0.5 * sigma ** 2) * delta_t + sigma * np.sqrt(delta_t) * Z_m)
    return S
```

```python
def Euler_aprox_GBM(S0, M, T, sigma, r):
    """Returns an euler aproximation of geometric brownian timeseries with volatility sigma of M + 1 steps over a period of T given a starting stock price S0, according to thr analaytical aproximation given in assignment 1 appendix C."""
    delta_t = T/M
    S = np.zeros(M+1)
    S[0] = S0
    np.random.seed(0)
    for t in range(1,M+1):
        Z_m = norm.rvs(0,1)
        S[t] = S[t-1] + r*S[t-1]*delta_t + sigma*S[t-1]*np.sqrt(delta_t)*Z_m
    return S
```

```python
# Parameters
S0 = 100  # Initial stock price
M = 1000  # Number of steps
T = 1     # Time horizon
sigma = 0.2  # Volatility
r = 0.05  # Risk-free rate

# Run both simulations
exact_path = exact_GBM(S0, M, T, sigma, r)
euler_path_M_1000 = Euler_aprox_GBM(S0=S0, M=1000, T=T, sigma=sigma, r=r)
mse = np.mean((exact_path - euler_path_M_1000)**2)
print('euler_path_M_1000 vs. analytical', mse)

# Plotting the results
plt.figure(figsize=(14, 7))
plt.plot(exact_path, label="Exact GBM")
plt.plot(euler_path_M_1000, label="Euler Approximation GBM", linestyle='--')
plt.title("Exact vs Euler Approximation GBM Simulation")
plt.xlabel("Time Steps")
plt.ylabel("Stock Price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
def Deltas_black_Scholes(price_timeseries, K, r, T, vol, num_days=7, mode='call'):
    price_timeseries = np.array(price_timeseries)
    num_steps = price_timeseries.shape[0]
    # Calculate deltas at specified intervals
    sparse_deltas = [BlackScholes(St=price, K=K, r=r, T=T, t=index / num_steps * T, vol=vol, mode=mode)[1] for index, price in enumerate(price_timeseries[::num_days])]
    # Initialize a full list of deltas with zeros
    full_deltas = np.zeros(num_steps)
    # Populate the full list of deltas by repeating each sparse delta value for 'num_days' times
    for i, delta in enumerate(sparse_deltas):
        full_deltas[i * num_days: (i + 1) * num_days] = delta
    # Ensure the last calculated delta is extended to the end of the array
    full_deltas[(i + 1) * num_days:] = sparse_deltas[-1]
    return full_deltas
```
